```python
# Phrease modelling
from gensim.models.phrases import Phrases
from gensim.models.word2vec import LineSentence
```

```python
def punct_space(token):
    return token.is_punct or token.is_space

def line_review(filename):
    with codecs.open(filename, encoding='utf_8') as f:
        for review in f:
            yield review.replace('\\n', '\n')
            
def lemmatized_sentence_corpus(filename):
    for parsed_review in nlp.pipe(line_review(filename), batch_size=1000, n_threads=2):
        for sent in parsed_review.sents:
            yield u' '.join([token.lemma_ for token in sent if not punct_space(token)])
```

```python
bigram_model_filepath = os.path.join('.', 'bigram_model_all')
trigram_model_filepath = os.path.join('.', 'trigram_model_all')
```

```python
bigram_model = Phrases.load(bigram_model_filepath)
trigram_model = Phrases.load(trigram_model_filepath)
```

```python
%%time

with codecs.open(trigram_review_filepath, 'w', encoding='utf_8') as f:
    for parsed_review in nlp.pipe(line_review(review_text_path), batch_size=1000, n_threads=2):
        unigram_review = [token.lemma_ for token in parsed_review if not punct_space(token)]
        bigram_review = bigram_model[unigram_review]
        trigram_review = trigram_model[bigram_review]
        trigram_review = [item for item in trigram_review if item not in spacy.lang.en.STOP_WORDS]
        
        trigram_review = u' '.join(trigram_review)
        f.write(trigram_review + '\n')
```

```python
# Topic Modelling

from gensim.corpora import Dictionary, MmCorpus
from gensim.models.ldamulticore import LdaMulticore

import pyLDAvis
import pyLDAvis.gensim
import warnings
import _pickle as pickle
```

```python
trigram_dictionary_filepath = os.path.join('.', 'trigram_dictionary.dict')
```

```python
%%time

trigram_review = LineSentence(trigram_review_filepath)
trigram_dictionary = Dictionary(trigram_review)
trigram_dictionary.filter_extremes(no_below=8, no_above=0.4, keep_n=10000)
trigram_dictionary.compactify()
trigram_dictionary.save(trigram_dictionary_filepath)
```

```python
def trigram_bow_generator(filepath):
    for review in LineSentence(filepath):
        yield trigram_dictionary.doc2bow(review)
```

```python
trigram_bow_filepath = os.path.join('.', 'trigram_bow_corpus_all.mm')
```

```python
%%time

MmCorpus.serialize(trigram_bow_filepath, trigram_bow_generator(trigram_review_filepath))
trigram_bow_corpus = MmCorpus(trigram_bow_filepath)
```

```python
lda_model_filepath = os.path.join('.', 'lda_model_all')
```

```python
%%time

with warnings.catch_warnings():
    warnings.simplefilter('ignore')
    lda = LdaMulticore(trigram_bow_corpus, num_topics=20, id2word=trigram_dictionary)
```

```python
lda.save(lda_model_filepath)
```

```python
def topic_test(topic_number, topn=25):
    print(u'{:20}{}'.format(u'term', u'frequency') +u'\n')
    for term, frequency in lda.show_topic(topic_number, topn=25):
          print(u'{:20}{:.3f}'.format(term, round(frequency,3)))
```

```python
topic_test(topic_number=1)
```

```python
LDAvis_data_filepath = os.path.join('.', 'ldavis_prepared')
```

```python
%%time

LDAvis_prepared = pyLDAvis.gensim.prepare(lda, trigram_bow_corpus, trigram_dictionary)

with open(LDAvis_data_filepath, 'wb') as f:
    pickle.dump(LDAvis_prepared, f)
    
# with open(LDAvis_data_filepath) as f:
#     LDAvis_prepared = pickle.load(f)
```
